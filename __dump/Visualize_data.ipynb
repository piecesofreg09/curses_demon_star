```python
import seaborn as sns
import numpy as np
import pandas as pd
import pickle, os
```

## try on short data

```python
data_dir = os.path.join(os.curdir, 'Data', 'basic_data_short.pkl')
with open(data_dir, 'rb') as in_file:
    ot = pickle.load(in_file)
data = ot['data']
target = ot['target']
data.shape
```

```python
sns.scatterplot(data=pd.DataFrame(-data.iloc[26, :]), legend=False)
sns.scatterplot(data=pd.DataFrame(-data.iloc[27, :]), legend=False)
sns.scatterplot(data=pd.DataFrame(-data.iloc[28, :]), legend=False)
sns.scatterplot(data=pd.DataFrame(-data.iloc[29, :]), legend=False)
```

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
Writer = animation.writers['ffmpeg']
writer = Writer(fps=20, metadata=dict(artist='Me'), bitrate=1800)
```

```python
fig = plt.figure(figsize=(10,6))

def animate(i):
    plt.clf()
    global data
    data_1 = pd.DataFrame(-data.iloc[i, :]) #select data range
    x1 = sns.scatterplot(data=data_1, legend=False)
    x1.text(1, 1, str(i))
    plt.ylim(0, 50)
    
    
ani = matplotlib.animation.FuncAnimation(fig, animate, frames=100, repeat=True)
ani.save('./Videos/test.mp4', writer=writer)
```

## Trying out a video output process with random numbers

```python
fig_1 = plt.figure(figsize=(10,7))
def animate_1(i):
    plt.clf()
    data = pd.DataFrame(np.random.rand(10, 1))
    x1 = sns.scatterplot(data=data, legend=False)
    x1.text(1, 1, str(i))
    
ani_1 = matplotlib.animation.FuncAnimation(fig_1, animate_1, frames=20, repeat=True)
ani_1.save('./Videos/test_1.mp4', writer=writer)
```
